```python
import re
```

```python
# Open File
log_file = open('model_s_001.txt', 'r')
```

```python
try:
    # Read all log data
    log_content = log_file.read()
finally:
    log_file.close()
```

```python
# Regular expression
pattern = re.compile(r"Mean AP:  [0-9]+.[0-9]+")
result = pattern.findall(log_content)

# Split from result
mean_ap = []
for item in result:
    mean_ap.append(float(item.split(":  ")[1]))
```

```python
import matplotlib.pyplot as plt

plt.plot(mean_ap)
plt.show()
```
